```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
plt.bar(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
plt.bar(x,y,color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data with outliers
np.random.seed(42)
data = np.random.normal(50, 10, 100)  # mean=50, std=10, 100 values
data = np.append(data, [90, 5])  # add outliers

# Create boxplot
plt.figure(figsize=(6, 6))
plt.boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor="skyblue"))

# Labels
plt.title("Boxplot Example with Median, Spread, and Outliers")
plt.ylabel("Values")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y1 = np.array([3, 7, 8, 5, 9])
y2 = np.array([1, 6, 4, 7, 2])
errors = np.array([0.5, 0.8, 0.6, 0.4, 0.7])

# Create subplots (2 rows, 3 columns)
fig, axs = plt.subplots(2, 3, figsize=(14, 8))

# 1. Pie chart
axs[0, 0].pie([40, 30, 20, 10], labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%', startangle=90)
axs[0, 0].set_title("Pie Chart")

# 2. Horizontal bar chart
axs[0, 1].barh(x, y1, color='skyblue')
axs[0, 1].set_title("Horizontal Bar Chart")
axs[0, 1].set_xlabel("Values")

# 3. Errorbar chart
axs[0, 2].errorbar(x, y1, yerr=errors, fmt='o-', capsize=5, color='red', label="Data with Error")
axs[0, 2].set_title("Error Bar Chart")
axs[0, 2].legend()

# 4. Area chart
axs[1, 0].fill_between(x, y1, color="lightgreen", alpha=0.7, label="Area Under Curve")
axs[1, 0].plot(x, y1, color="green")
axs[1, 0].set_title("Area Chart")
axs[1, 0].legend()

# 5. Line chart with grid
axs[1, 1].plot(x, y1, label="y1", marker='o')
axs[1, 1].plot(x, y2, label="y2", marker='s')
axs[1, 1].set_title("Line Chart with Grid & Legend")
axs[1, 1].legend()
axs[1, 1].grid(True)

# 6. Empty (you can add any chart)
axs[1, 2].text(0.3, 0.5, "Add Your Own Chart!", fontsize=12)
axs[1, 2].set_axis_off()

plt.tight_layout()
plt.show()
```

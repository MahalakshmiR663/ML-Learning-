```python
sns.relplot(data=tips,x='total_bill',y='tip',col='day',col_wrap=2,hue='smoker',size='size')

#hue is like saying "color by this column".
```

```python
sns.relplot(
    data=tips,
    x='total_bill', 
    y='tip',
    col='time',
    row='sex',          # subplot rows by sex
    hue='smoker',       # color by smoker
    style='sex',        # marker style by sex    .male,*female
    size='size',        # size by party size
    kind='scatter',     # can change to 'line'
    palette='coolwarm', # custom colors
    sizes=(20, 200),    # control min/max size
          # wrap columns into rows if >2
    height=4, aspect=1  # subplot size control
)
"""row / col in Seaborn faceting (FacetGrid) tell Seaborn how to split the data and arrange subplots.

row="..."  each unique category gets a subplot stacked vertically.||||

col="..."  each unique category gets a subplot arranged horizontally.------

col_wrap is only for wrapping columns when you use col.

Example: if you have 6 categories and col_wrap=3, Seaborn will make a grid with 2 rows  3 columns."""
```

```python
'''hue = color

size = marker size

style = marker shape

row = subplot rows

col = subplot columns

row and col_wrap can't be used together bcz row means vertically and col_wrap for vertically so its get confused which one to take up'''
```

```python
# Facet by rows and columns
sns.relplot(data=tips, x="total_bill", y="tip", col="time", row="sex",hue="smoker",style="sex")

# Facet by columns with wrapping
sns.relplot(data=tips, x="total_bill", y="tip", col="day", col_wrap=2,hue="sex")
#col_wrap basically decides no of plots in a line if its 2 then 2 plots horizontally placed if 3 then 3 plots r placed horizontally
```

```python
sns.relplot(data=iris,x='sepal_length',y='petal_length',col='species',hue='species')
```

```python
sns.relplot(data=iris,x='sepal_length',y='petal_length',row='species',hue='species')
```

```python
'''x, y  define the main axes of the plot (what goes on the x-axis and y-axis).

row, col  define how the dataset is split into subplots (faceting).

hue  defines color grouping within each subplot.

 So in short:

Main plot content = x and y.

Arrangement into multiple plots = row and col.'''
```
